```python
from peleffybenchmarktools.minimization.benchmark import MinimizationBenchmark
```

# OpenFF Optimization Set 1

## Normal_OpenFF:
- openff_unconstrained-1.2.1
- am1bcc
- optimized

```python
benchmark_set1 = MinimizationBenchmark(dataset_name='OpenFF Optimization Set 1',
                                       output_path='normal_OpenFF',
                                       n_proc=32,
                                       geometry_selection='optimized',
                                       PELE_exec='/home/municoy/builds/PELE/PELE-repo_serial/PELE-1.6',
                                       PELE_src='/home/municoy/repos/PELE-repo/',
                                       PELE_license='/home/municoy/builds/PELE/license',
                                       forcefield='openff_unconstrained-1.2.1.offxml',
                                       charge_method='am1bcc',
                                       force_parameterization=True)
```

```python
ref_pdb_paths1 = benchmark_set1.retrieve_reference_pdbs(filter_dihedrals=False)
```

```python
from glob import glob

ref_pdb_paths1 = list(glob('normal_OpenFF/QM/*.pdb'))
```

```python
min_pdb_paths1 = benchmark_set1.minimize_structures(ref_pdb_paths1)
```

```python
from glob import glob

min_pdb_paths1 = list(glob('normal_OpenFF/PELE/*/minimized.pdb'))
```

```python
benchmark_set1.compute_RMSD(min_pdb_paths1,
                            ref_pdb_paths1,
                            labeling1='folder',
                            labeling2='file')
```

```python
benchmark_set1.energetic_difference(min_pdb_paths1)
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_bond_differences(ref_pdb_paths1,
                                        min_pdb_paths1,
                                        labeling1='file',
                                        labeling2='folder')
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_torsion_differences(ref_pdb_paths1,
                                           min_pdb_paths1,
                                           labeling1='file',
                                           labeling2='folder')
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_dihedral_differences(ref_pdb_paths1,
                                            min_pdb_paths1,
                                            labeling1='file',
                                            labeling2='folder')
```
