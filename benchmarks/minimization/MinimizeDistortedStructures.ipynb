# OpenFF Optimization Set 1

## distorted_OpenFF_1:
- openff_unconstrained-1.2.1
- am1bcc
- distorted

```python
benchmark_set1 = MinimizationBenchmark(dataset_name='OpenFF Optimization Set 1',
                                       output_path='distorted_OpenFF_1',
                                       n_proc=32,
                                       PELE_exec='/home/municoy/builds/PELE/PELE-repo_serial/PELE-1.6',
                                       PELE_src='/home/municoy/repos/PELE-repo/',
                                       PELE_license='/home/municoy/builds/PELE/license',
                                       forcefield='openff_unconstrained-1.2.1.offxml',
                                       charge_method='am1bcc',
                                       distort_dihedrals=True,
                                       distort_value_for_dihedrals=10,
                                       distort_bonds=True,
                                       distort_value_for_bonds=0.5,
                                       distort_torsions=True,
                                       distort_value_for_torsions=20,
                                       random_distort=False,
                                       force_parameterization=True)
```

```python
ref_pdb_paths1 = benchmark_set1.retrieve_reference_pdbs(filter_dihedrals=False)
```

```python
from glob import glob

ref_pdb_paths1 = list(glob('distorted_OpenFF_1/QM/*.pdb'))
```

```python
min_pdb_paths1 = benchmark_set1.minimize_structures(ref_pdb_paths1)
```

```python
from glob import glob

dist_pdb_paths1 = list(glob('distorted_OpenFF_1/PELE/*/ligand.pdb'))
```

```python
from glob import glob

min_pdb_paths1 = list(glob('distorted_OpenFF_1/PELE/*/minimized.pdb'))
```

```python
benchmark_set1.compute_RMSD(dist_pdb_paths1,
                            ref_pdb_paths1,
                            labeling1='folder',
                            labeling2='file')
```

```python
benchmark_set1.compute_RMSD(min_pdb_paths1,
                            ref_pdb_paths1,
                            labeling1='folder',
                            labeling2='file')
```

```python
benchmark_set1.energetic_difference(min_pdb_paths1)
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_bond_differences(ref_pdb_paths1,
                                        dist_pdb_paths1,
                                        labeling1='file',
                                        labeling2='folder',
                                        range=(0, 2))
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_bond_differences(ref_pdb_paths1,
                                        min_pdb_paths1,
                                        labeling1='file',
                                        labeling2='folder',
                                        range=(0, 2))
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_torsion_differences(ref_pdb_paths1,
                                           dist_pdb_paths1,
                                           labeling1='file',
                                           labeling2='folder',
                                           range=(0, 20))
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_torsion_differences(ref_pdb_paths1,
                                           min_pdb_paths1,
                                           labeling1='file',
                                           labeling2='folder',
                                           range=(0, 20))
```

```python
from peleffy.utils import Logger
logger = Logger()
logger.set_level('CRITICAL')

benchmark_set1.compute_dihedral_differences(ref_pdb_paths1,
                                            dist_pdb_paths1,
                                            labeling1='file',
                                            labeling2='folder',
                                            range=(0, 20))
```
