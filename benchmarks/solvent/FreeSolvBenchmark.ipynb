# Hydration energies benchmark

```python
from peleffybenchmarktools.solvent import SolventBenchmark
```

```python
PELE_EXEC = '/home/municoy/builds/PELE/PELE-repo_serial/PELE-1.6'
PELE_SRC =  '/home/municoy/repos/PELE-repo/'
PELE_LICENSE = '/home/municoy/builds/PELE/license'
SCHRODINGER_SRC = '/data/general_software/schrodinger2017-2/'
PLOPROTTEMP_SRC = '/home/municoy/repos/PlopRotTemp/main.py'
```

4. PELE with `OFF` parameters, the `OBC` implicit solvent and `am1bcc` charges
5. PELE with `OFF` parameters, the `OBC` implicit solvent and `OPLS` charges
6. PELE with `OFF` bonding parameters, the `OPLS2005` non-bonding parameters, `VDGBNP` implicit solvent and `OPLS` charges
7. PELE with `OFF` bonding parameters, the `OPLS2005` non-bonding parameters, `OBC` implicit solvent and `OPLS` charges
8. PELE with `OFF` dihedral parameters, the `OPLS2005` non-bonding, bond and angle parameters, `VDGBNP` implicit solvent and `OPLS` charges.
9. PELE with `OFF` dihedral parameters, the `OPLS2005` non-bonding, bond and angle parameters, `OBC` implicit solvent and `OPLS` charges.
10. PELE with `OFF` bonding parameters, the `OPLS2005` non-bonding parameters, `VDGBNP` implicit solvent and `am1bcc` charges.
11. PELE with `OFF` dihedral parameters, the `OPLS2005` non-bonding, bond and angle parameters, `VDGBNP` implicit solvent and `am1bcc` charges.
12. PELE with `OFF` dihedral parameters, the `OPLS2005` non-bonding, bond and angle parameters, `OBC` implicit solvent and `am1bcc` charges.

## 1. Standard PELE with `OPLS2005` parameters and the `VDGBNP` implicit solvent

```python
# Initialize the solvent benchmark
benchmark = SolventBenchmark(pele_exec=PELE_EXEC,
                             pele_src=PELE_SRC,
                             pele_license=PELE_LICENSE,
                             ploprottemp_src=PLOPROTTEMP_SRC,
                             schrodinger_src=SCHRODINGER_SRC,
                             forcefield_name='OPLS2005',
                             charge_method='OPLS',
                             solvent='VDGBNP',
                             n_proc=20)
```

```python
# Run the benchmark
benchmark.run('OFF_out1')
```

```python
# Plot the results
benchmark.plot_results()
```

```python
benchmark.display_outlayers_above(10)
```

```python
benchmark.to_csv('OFF_out1/results.csv')
```
